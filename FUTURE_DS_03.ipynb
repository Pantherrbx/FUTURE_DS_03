```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and Clean Data
df = pd.read_csv('student_feedback.csv')

# Drop ID columns not needed for analysis
df_clean = df.drop(columns=['Unnamed: 0', 'Student ID'])

# 2. Calculate Descriptive Statistics
# Calculate mean rating for each question
avg_ratings = df_clean.mean().sort_values(ascending=False)

# 3. Visualization

# Set visual style
sns.set(style="whitegrid")

# Plot A: Average Ratings Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette="viridis")
plt.title('Average Feedback Ratings by Category', fontsize=16)
plt.xlabel('Average Rating (1-10)', fontsize=12)
plt.xlim(0, 10)  # Fix scale to max rating
plt.show()

# Plot B: Correlation Heatmap
plt.figure(figsize=(10, 8))
# Calculate correlation matrix
corr_matrix = df_clean.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Feedback Metrics', fontsize=16)
plt.show()

# Plot C: Distribution of Recommendations
plt.figure(figsize=(8, 6))
sns.histplot(df_clean['Course recommendation based on relevance'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Course Recommendations', fontsize=16)
plt.xlabel('Rating (1-10)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# 4. Save Cleaned Data
df_clean.to_csv('cleaned_student_feedback.csv', index=False)
print("Analysis Complete. Cleaned data saved.")
```
